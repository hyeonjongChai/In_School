```python
def pearsonr(X, Y):
    # print(X, Y)
    meanX = sum(X) / len(X)
    meanY = sum(Y) / len(X)
    # print(f"mean X: {meanX}, mean Y: {meanY}")

    stdX = (sum((x - meanX)**2 for x in X)/len(X)) ** 0.5
    stdY = (sum((y - meanY)**2 for y in Y)/len(Y)) ** 0.5
    # print(f"std X: {stdX}, std Y: {stdY}")

    cov = sum((x-meanX)*(y-meanY) for x, y in zip(X, Y)) / len(X) 
    corr = cov / stdX / stdY
    # print(f"cov: {cov}, corr: {corr}")
    print(f"Pearson Correlation Coefficient: {corr}")
```

```python
idx_HDI = list(df7['human_development_index'].index)
idx_VAC = list(df7['people_vaccinated_per_hundred'].index)

X = []
Y = []

for hdi, vac in zip(idx_HDI, idx_VAC):
    X.append(df7['human_development_index'][hdi])
    Y.append(df7['people_vaccinated_per_hundred'][vac])

print(X)
print(Y)
```

```python
import torch

# y = WX + b
x_train = torch.FloatTensor(X)
y_train = torch.FloatTensor(Y)

x_train = x_train.view(-1, 1)
y_train = y_train.view(-1, 1)

W = torch.zeros(1,1)
b = torch.zeros(1,1)

lr = 0.3

minCost = 100000
expW = 0
expB = 0
Epoch = 10001

for epoch in range(Epoch):
    W.requires_grad_(True)
    b.requires_grad_(True)

    hypothesis = torch.mm(x_train, W) + b
    cost = torch.mean((hypothesis - y_train)**2)

    cost.backward()

    with torch.no_grad() as grd:
        W = W - lr * W.grad
        b = b - lr * b.grad

        if minCost > cost.item():
            minCost = cost.item()
            expW = W
            expB = b
        

        if epoch % 100 == 0:
            print(f'epoch: {epoch}, cost: {cost.item():.4f}, W: {W}, b: {b}')

print(f"\n\nlerning rate: {lr}, epoch: {Epoch}, min_cost: {minCost}, W: {expW}, b: {expB}")

# lerning rate: 0.01, epoch: 10001, min_cost: 245.6505126953125, W: tensor([[143.5373]]), b: tensor([[-61.8593]])
# lerning rate: 0.05, epoch: 10001, min_cost: 244.6042022705078, W: tensor([[150.3016]], requires_grad=True), b: tensor([[-66.8402]], requires_grad=True)
# lerning rate: 0.3, epoch: 10001, min_cost: 244.6042022705078, W: tensor([[150.3173]], requires_grad=True), b: tensor([[-66.8517]], requires_grad=True)
# lerning rate: 0.4, epoch: 10001, min_cost: 244.60421752929688, W: tensor([[150.2959]], requires_grad=True), b: tensor([[-66.8359]], requires_grad=True)
```
