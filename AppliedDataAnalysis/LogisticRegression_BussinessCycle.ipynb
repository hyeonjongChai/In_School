```python
df = pd.read_csv('C:/Users/yssg9187/Desktop/20-2/3. 응용자료분석/D_data/MACRO_202006.csv')
```

```python
df['QUARTER'] = ((df['YYM'] % 100) / 3).astype(int)
df['RollingMean']= df.REALGDP.rolling(20).mean()
df['TARGET'] = (df.REALGDP > df.RollingMean).astype(int)
df['TARGET1'] = df['TARGET'].shift(-1) 
pct_cols = ['M2', 'INFL']
df.loc[:, pct_cols] = df.loc[:, pct_cols].pct_change(1)
df.TARGET1.value_counts()
df1 = pd.get_dummies(df, columns=['QUARTER'])
df1 = pd.get_dummies(df, columns=['QUARTER'], drop_first=True)
df1 = pd.get_dummies(df, columns=['QUARTER'], drop_first=True).dropna()
df2 = df1.drop(['YYM', 'REALGDP', 'TARGET', 'RollingMean'], axis = 1)
```

실물변수('REALCONS', 'INV', 'UNEMP', 'EMPLOY), 분기더미('QUARTER_2', 'QUARTER_3', 'QUARTER_4) 데이터를 가지고 1분기후의 호황여부를 Machine Learning을 이용하여 예측하고자 합니다.

```python
df1['const'] = 1
x =df1[['const', 'REALCONS', 'INV', 'UNEMP', 'EMPLOY', 'QUARTER_2', 'QUARTER_3', 'QUARTER_4']].to_numpy()
```

```python
y =df1['TARGET1'].to_numpy()
```

1. x_data와 y_data를 추출합니다.

```python
x_data = x[:, 1:] #const 상수 제외
```

```python
y_data = y 
```

```python
x_data = np.asarray(x_data, dtype=np.float32)
y_data = np.asarray(y_data, dtype=np.float32)
```

2. tensorflow data API를 이용하여 dataset을 구성합니다.

```python
dataset = tf.data.Dataset.from_tensor_slices((x_data, y_data)).batch(len(x_data))
```

```python
k = x_data.shape[1]
```

3. W, b를 초기화시킵니다.

```python
W = tf.Variable(tf.random.normal(([k, 1])), name='weight')
b = tf.Variable(tf.random.normal(([1])), name='bias')
```

4. Sigmoid 함수를 가설로 선언합니다.

```python
def logistic_regression(features):
    hypothesis = tf.math.divide(1., 1+tf.math.exp(tf.matmul(features,W) + b)) 
    return hypothesis
```

5. 가설을 검증할 Cost 함수를 정의합니다.

```python
def loss_fn(hypothesis, labels):
    cost = - tf.reduce_mean(y * tf.math.log(hypothesis) + (1-y) * tf.math.log(1-hypothesis))
    return cost
```

6. gradient descent algorithm을 이용하여 cost함수를 최소화시키는 과정을 함수를 정의합니다

```python
optimizer = tf.optimizers.SGD(learning_rate=0.005)# Stochastic Gradient Descent algorithm
```

```python
def grad(features, labels):
    with tf.GradientTape() as tape:
        loss_value = loss_fn(logistic_regression(features), labels)
    return tape.gradient(loss_value, [W,b])
```

8. 5001 반복시행하여 500번에 해당될때 cost, accuray를 보입니다.

```python
EPOCHS = 5001

for step in range(EPOCHS):
    for features, labels  in iter(dataset):
        
        hypothesis = logistic_regression(features)
        grads = grad(features, labels)
        optimizer.apply_gradients(grads_and_vars=zip(grads,[W,b]))
        cost = loss_fn(hypothesis,labels)
        
        predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
      
        accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, labels), dtype=tf.float32))
        
        if step % 100 == 0:
            print("Iter: {}, Loss: {:.4f},  Accuracy: {:.4f}".format(step, cost, accuracy))
```
